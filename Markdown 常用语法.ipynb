# 字体
- 加粗
    
    用两个* 包裹内容
    **加粗字体**
    
- 斜体
   
    用一个* 包裹内容
    *倾斜字体*
   
- 斜体加粗

    用三个* 包裹内容
    ***加粗并倾斜***

- 删除线

    用两个~~ 包裹内容
    ~~删除线~~

# 图片
语法：
    !+"[图片alt]"+"(图片地址 ''图片title'')"

    
# 超链接
语法：
    "[超链接名]" + "(超链接地址 "超链接title")"
   
# 列表
## -无序列表
   语法： 用-、+、* 任意一种即可
## -有序列表
   语法： 数字加点
## -列表嵌套
   语法：上级和下级之间敲三个空格即可
   * 一级无序列表
      * 二级无序列表
      
   2. 一级有序列表
      1. 二级有序列表
      
# 表格
语法：

表头|表头|表头

---|:--:|---:

内容|内容|内容

内容|内容|内容

第二行分割表头和内容
-有一个就行了，为了对齐可以多加几个
文字默认居左
-两边加：文字居中
-右边加：文字居右

# 代码
语法：

单行代码之间分别用一个反引号包起来

`代码内容`

代码块：用三个反引号包起来，且两边反引号都要单独占一行

```
代码内容
代码内容
```
