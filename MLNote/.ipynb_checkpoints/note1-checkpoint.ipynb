# 1 什么是机器学习
利用计算机从历史数据中找出规律，并把这些规律用到对未来不确定场景的决策

## 1.1 从数据中寻找规律
### 1.1.1 传统统计学处理数据方式

抽样->描述统计->结论->假设检验


### 1.1.2 ML
ML就不需要抽样，而是直接分析所有数据

用模型刻画（拟合）规律，找合适的函数，即寻找对应模型

## 1.2 机器学习发展的原动力
从历史数据中找出规律，把这些规律用到对未来自动做出决定。

用数据代替expert

经济驱动，数据变现

### 1.2.1 业务系统发展历史
1. 基于专家经验（运维和产品头脑风暴得出业务逻辑，程序员用一堆if else实现）
2. 基于统计-- 分维度统计。数据分析，受限于数据分析人员的经验（数据报表，联机事务处理OLAPP（on-line transaction processing））
3. 机器学习：
   - 离线机器学习（每天定时更新，跑算法，来训练生成一个新模型，一直循环）
   - 在线学习（实时手机网络上的数据，进行分析，会覆盖之前的数据）

# 3 机器学习和数据分析的区别

- 处理数据类型： 交易数据 vs 行为数据

传统数据分析处理的是**交易数据**（跟钱有关的数据）

机器学习处理的是**行为数据**（搜索历史、点击历史、浏览历史、评论）

- 数据量： 少量数据 vs 海量数据

- 分析方法： 采样分析 vs 全量分析

- 解决业务问题不同

OLAP是报告过去发生的事情，机器学习是预测未来的事情



- 技术手段不同
OLAP：用户驱动、交互式分析
数据挖掘：数据驱动、自动进行知识发现

- 参与者不同

    分析师：分析师能力决定结果（服务目标用户：公司高层）
    数据+算法：数据质量决定结果（服务目标用户：个体）
    
# 4 机器学习算法分类
## 算法分类（1）
- 有监督学习
训练数据已经被提前打好了标签
   - 分类算法
   - 回归算法

- 无监督学习
全部无标签
   - 聚类


- 半监督学习（强化学习）
部分有标签

## 算法分类（2）
根据解决的问题来分类
- 分类和回归
- 聚类
- 标注

## 算法分类（3）很重要
直指算法本质
- 生成模型
不会明确的告诉结果Y，而是给出可能各个结果的概率
- 判别模型
给一个X，返回一个明确的Y（非1即2的感觉）

## 常见算法一览

- FP-Growth（华人发明）
- 逻辑回归（百度谷歌搜索结果的排序）
- RF、GBDT（决策树算法的改进）
- 推荐算法（电商网站标配）
- LDA（文本分析、自然语言处理）
- Word2Vector（文本挖掘）
- HMM、CRF（自然语言处理文本挖掘有关的）
- 深度学习（现在最火的）

# 机器学习解决问题

1. 确定目标
   - 业务需求（确定目标）
   - 数据（收集数据）
   - 特征工程（数据的处理，提取特征）很重要，影响最终产出，整体中占用70%的比重
2. 训练模型
   - 定义模型（就是一个公式函数）
   - 定义损失函数（定义预测结果与真实结果的相似程度的一个评价）
   - 优化算法（梯度下降、随机梯度下降、坐标法等）
3. 模型评估
   - 交叉验证
   - 效果评估
