# 基于内容的推荐系统的原理

需要三个东西
1. 电影评分表 S
2. 电影内容矩阵X（就是电影的future）
3. Theta矩阵（用户对每种电影喜好程度的权重）

***两个已知：电影评分表S、电影内容矩阵X，来求Theta矩阵***

根据每部电影的内容X以及已经评过分的电影S来推断出每个用户对于每部电影的喜好程度Theta。然后再根据得到的Thate加上电影的特征X来预测出每个用户没有看过的电影的评分。然后将评分从高到低的顺序排列来推荐给用户。
