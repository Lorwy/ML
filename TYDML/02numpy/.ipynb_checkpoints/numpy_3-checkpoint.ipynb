```python
A = np.array([[1,1],
             [0,1]])
B = np.array([[2,0],
             [3,4]])
print(A)
print('-------------')
print(B)
print('-------------')
print(A*B) # 对应位置的相乘，感觉是点乘
print('-------------')
print(A.dot(B)) # 矩阵的乘法
print('-------------')
print(np.dot(A,B)) # 矩阵的乘法
```
